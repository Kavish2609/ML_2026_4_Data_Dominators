```python
import os
import shutil
import random

# Configuration
source_folder = 'training_data1_v2'
subsets = ['train', 'val', 'test']
ratios = [0.75, 0.15, 0.10]

# 1. Create the destination folders in the current root
for folder in subsets:
    if not os.path.exists(folder):
        os.makedirs(folder)
        print(f"Created folder: {folder}")

# 2. Identify and shuffle all subject directories
# This looks inside 'training_data1_v2' for the BraTS folders
subjects = [s for s in os.listdir(source_folder) if os.path.isdir(os.path.join(source_folder, s))]
random.seed(42)
random.shuffle(subjects)

# 3. Calculate split indices
total = len(subjects)
train_end = int(total * ratios[0])
val_end = train_end + int(total * ratios[1])

# 4. Map subjects to their new homes
split_map = {
    'train': subjects[:train_end],
    'val': subjects[train_end:val_end],
    'test': subjects[val_end:]
}

# 5. Move the data
print("\nStarting move...")
for folder_name, folder_list in split_map.items():
    for subject in folder_list:
        src = os.path.join(source_folder, subject)
        dst = os.path.join(folder_name, subject)
        shutil.move(src, dst)

print(f"\nSuccessfully moved {total} folders.")
print(f"Distribution: {len(split_map['train'])} Train, {len(split_map['val'])} Val, {len(split_map['test'])} Test")
```
